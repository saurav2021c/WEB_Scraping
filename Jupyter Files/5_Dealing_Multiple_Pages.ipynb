```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.airbnb.co.in/s/Mumbai--Maharashtra/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&monthly_start_date=2023-08-01&monthly_length=3&price_filter_input_type=0&price_filter_num_nights=2&channel=EXPLORE&search_type=user_map_move&query=Mumbai%2C%20Maharashtra&place_id=ChIJwe1EZjDG5zsRaYxkjY_tpF0&date_picker_type=calendar&checkin=2023-07-05&checkout=2023-07-07&source=structured_search_input_header&ne_lat=19.822246668822423&ne_lng=74.92855971147623&sw_lat=17.052300999722576&sw_lng=71.97261953194942&zoom=8&zoom_level=8&search_by_map=true&location_search=MIN_MAP_BOUNDS'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
df = pd.DataFrame({'Links':[''], 'Title': [''], 'Details':[''], 'Price':[''], 'Ratings':['']})
```

```python
# To Extract data from a single page
postings = soup.find_all('div', class_ = 'c4mnd7m dir dir-ltr')
for post in postings:
    try:
        link = post.find('a', class_ = 'l1ovpqvx bn2bl2p dir dir-ltr').get('href')
        link_full = 'https://www.airbnb.co.in'+ link
        title = post.find('div', class_ = 't1jojoys dir dir-ltr').text
        price = post.find('span', class_ = 'a8jt5op dir dir-ltr').text
        rating = post.find('span', class_ = 'r1dxllyb dir dir-ltr').text
        details = post.find('span', class_ = 'dir dir-ltr').text
        df = df.append({'Links': link_full, 'Title': title, 'Details':details, 'Price': price, 'Ratings':rating}, ignore_index = True)
    except:
        pass
print("Process Completed")
```

```python
while True:
    postings = soup.find_all('div', class_ = 'c4mnd7m dir dir-ltr')
    for post in postings:
        try:
            link = post.find('a', class_ = 'l1ovpqvx bn2bl2p dir dir-ltr').get('href')
            link_full = 'https://www.airbnb.co.in'+ link
            title = post.find('div', class_ = 't1jojoys dir dir-ltr').text
            price = post.find('span', class_ = 'a8jt5op dir dir-ltr').text
            rating = post.find('span', class_ = 'r1dxllyb dir dir-ltr').text
            details = post.find('span', class_ = 'dir dir-ltr').text
            df = df.append({'Links': link_full, 'Title': title, 'Details':details, 'Price': price, 'Ratings':rating}, ignore_index = True)
        except:
            pass
    next_page = soup.find('a', {'aria-label':'Next'}).get('href')
    next_page_full = 'https://www.airbnb.co.in'+next_page
    url = next_page_full
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')
print("Process Completed")
```
