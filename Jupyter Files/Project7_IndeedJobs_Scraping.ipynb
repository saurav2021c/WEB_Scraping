# This Code is automation of whole web Scraping Process

```python
#1 Starts the driver and goes to our starting webpage
driver = webdriver.Chrome()
driver.get('https://in.indeed.com/')
time.sleep(2)

#2 inputs What and Where
what = driver.find_element(By.XPATH, '//*[@id="text-input-what"]')
what.send_keys('Data Analyst')
where = driver.find_element(By.XPATH, '//*[@id="text-input-where"]')
where.send_keys('Mumbai,', 'Maharashtra')
FindButton = driver.find_element(By.XPATH, '//*[@id="jobsearch"]/button').click()
time.sleep(2)

#3 To store data in data Frame table
df = pd.DataFrame({'Links':[''], 'Company': [''], 'Job Title':[''], 'Rating':[''], 'Date Posted':['']})

#4 To find postings in a container by container on page 1 page
postings = soup.find_all('div', class_ = 'slider_container css-8xisqv eu4oa1w0')
page_count = 0

#5 This loop goes through every page and grabs all the details of each posting
while True:

    #6 Imports the HTML of the current page into python
    soup = BeautifulSoup(driver.page_source, 'lxml')
    #6 Grabs the HTML of each posting in current page
    postings = soup.find_all('div', class_ = 'slider_container css-8xisqv eu4oa1w0')
    
    #7 grabs all the details for each posting and adds it as a row to the dataframe
    for post in postings:
        link = post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').get('href')
        link_full = 'https://in.indeed.com'+ link
        title = post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').text.strip()
        company = post.find('span', class_ = 'companyName').text
        try:
            rating = post.find('span', class_ = 'ratingNumber').text
        except:
            rating = 0
        try:
            Date = post.find('span', class_ = 'date').text
        except:
            Date = 'NA'
        #8 Append the data of current page by container by container
        df = df.append({'Links':link_full, 'Company': company, 'Job Title':title, 'Rating':rating, 'Date Posted':Date}, ignore_index = True)
    
    page_count += 1
   
    #9 run till the given count of pages then it will stop the process
    if page_count>70:
        print("Process Completed Sucessfully")
        break
       
    #10 Grab the next page url
    try:
        button = soup.find('a', {'aria-label' : 'Next Page'}).get('href')
        driver.get('https://in.indeed.com'+button)
        time.sleep(1)
    except:
        break
    print('Your Page Count is: ',page_count)
   
    
```
